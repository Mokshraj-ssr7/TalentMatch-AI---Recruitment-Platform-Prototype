```python
# Step 2: Create Job Openings
jobs_data = {
    'job_id': ['JOB001', 'JOB002', 'JOB003'],
    'title': [
        'Backend Python Developer',
        'AI/ML Engineer',
        'Full Stack Developer'
    ],
    'description': [
        'We need a Backend Python Developer with strong Django/Flask experience to build REST APIs and microservices for our fintech platform. Must have 3+ years experience with databases and cloud deployment.',
        'Looking for AI/ML Engineer to develop NLP models and predictive analytics solutions. Required: Python, TensorFlow/PyTorch, experience with transformers and deep learning. 2+ years ML experience.',
        'Full Stack Developer role requiring React frontend and Node.js backend expertise. Will build customer-facing SaaS product. Need experience with MongoDB and modern JavaScript frameworks.'
    ],
    'required_experience': [3, 2, 3]
}

df_jobs = pd.DataFrame(jobs_data)
print("ðŸ’¼ Job Openings Created!")
print(f"Total Open Positions: {len(df_jobs)}\n")
for idx, row in df_jobs.iterrows():
    print(f"{row['job_id']}: {row['title']}")
    print(f"   â†’ {row['description'][:100]}...")
    print()
```

```python
# Step 3: Semantic Matching using TF-IDF (Simple, No External APIs)
# Note: In production, we'd use sentence-transformers for better accuracy
# This demo uses TF-IDF to show the concept without requiring external libraries

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def match_candidates_to_job(job_description, candidate_profiles, top_n=3):
    """
    Match candidates to job using semantic similarity
    Returns: DataFrame with top N matches and scores
    """
    # Combine job description with all candidate profiles
    all_texts = [job_description] + candidate_profiles.tolist()
    
    # Create TF-IDF vectors (captures semantic meaning beyond keywords)
    vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1, 2))
    tfidf_matrix = vectorizer.fit_transform(all_texts)
    
    # Calculate cosine similarity between job and each candidate
    job_vector = tfidf_matrix[0:1]
    candidate_vectors = tfidf_matrix[1:]
    similarities = cosine_similarity(job_vector, candidate_vectors)[0]
    
    return similarities

print("ðŸ¤– Semantic Matching Engine Ready!")
print("âœ… Using TF-IDF + Cosine Similarity (Production version would use transformer embeddings)")
print("ðŸ“Š This approach captures semantic meaning, not just keyword matches\n")
```

```python
# Step 4: Run Matching for All Jobs
results_all_jobs = []

for idx, job in df_jobs.iterrows():
    print(f"\n{'='*80}")
    print(f"ðŸŽ¯ MATCHING FOR: {job['title']} ({job['job_id']})")
    print(f"{'='*80}")
    
    # Get similarity scores
    scores = match_candidates_to_job(job['description'], df_candidates['profile'])
    
    # Create results dataframe
    results = df_candidates.copy()
    results['match_score'] = scores
    results['match_percentage'] = (scores * 100).round(1)
    results['job_id'] = job['job_id']
    results['job_title'] = job['title']
    
    # Sort by match score
    results = results.sort_values('match_score', ascending=False)
    
    # Display top 3 matches
    print(f"\nðŸ† TOP 3 CANDIDATES:\n")
    for i, (_, candidate) in enumerate(results.head(3).iterrows(), 1):
        print(f"#{i} {candidate['name']} - {candidate['match_percentage']}% Match")
        print(f"   Skills: {candidate['skills']}")
        print(f"   Experience: {candidate['experience_years']} years")
        print(f"   Why matched: {candidate['profile'][:120]}...")
        print()
    
    results_all_jobs.append(results)

# Combine all results
df_all_matches = pd.concat(results_all_jobs, ignore_index=True)
print(f"\nâœ… Matching Complete! Total match records: {len(df_all_matches)}")
```

```python
# Step 5: Visualize Matching Results
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Candidate-Job Match Scores (AI-Powered Semantic Matching)', fontsize=15, fontweight='bold')

for idx, job_id in enumerate(df_jobs['job_id']):
    job_title = df_jobs[df_jobs['job_id'] == job_id]['title'].values[0]
    job_matches = df_all_matches[df_all_matches['job_id'] == job_id].sort_values('match_score', ascending=True)
    
    # Create horizontal bar chart
    colors = ['#27ae60' if score >= 0.3 else '#f39c12' if score >= 0.2 else '#e74c3c' 
              for score in job_matches['match_score']]
    
    axes[idx].barh(job_matches['name'], job_matches['match_percentage'], color=colors)
    axes[idx].set_xlabel('Match Score (%)', fontweight='bold')
    axes[idx].set_title(f'{job_title}', fontweight='bold', fontsize=11)
    axes[idx].axvline(x=30, color='green', linestyle='--', alpha=0.5, label='Good Match (>30%)')
    axes[idx].grid(axis='x', alpha=0.3)
    axes[idx].set_xlim(0, max(job_matches['match_percentage']) + 10)
    
    # Add value labels
    for i, v in enumerate(job_matches['match_percentage']):
        axes[idx].text(v + 1, i, f'{v}%', va='center', fontsize=8, fontweight='bold')

plt.tight_layout()
plt.show()

print("\nðŸŽ¨ Visualization Key:")
print("  ðŸŸ¢ Green: Strong Match (>30%)")
print("  ðŸŸ  Orange: Moderate Match (20-30%)")
print("  ðŸ”´ Red: Weak Match (<20%)")
```
